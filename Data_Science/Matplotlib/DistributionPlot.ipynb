```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# We use ditribution plot to visualize the distribution of quantitive data
```

```python
df = sns.load_dataset('tips')
```

```python
df['size'].unique()
```

```python
sns.displot(df)
```

```python
sns.displot(df["total_bill"], bins = 5)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "kde")
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "hex")
```

# Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = "sex", palette= "rainbow")
```
