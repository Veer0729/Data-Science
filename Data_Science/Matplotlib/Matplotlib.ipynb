```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
x
y = x** 2
y
```

```python
plt.plot(x, y)
plt.title("title")
plt.xlabel("label 1")
plt.ylabel("label 2")
```

# Subplot in Matplotlib

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x, y)
plt.subplot(2,2,3)
plt.plot(y, x)
plt.subplot(2,2,2)
plt.plot(x**2, x**2)
```

# Object oriented way

```python
fig = plt.figure()
axis = fig.add_axes([0.1, 0.1, 1, 1])
axis.plot(x,y)
axis2 = fig.add_axes([0.3, 0.4, .3, 0.3])
axis2.plot(x,y)
```

# Types of Plot

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

# Saving plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y)
plt.savefig("basicplot.png")
```
